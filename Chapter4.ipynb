## 4-2 Discreteな行動

離散値を行動にする要素をもつ。
ある行動に１つの整数を割り当てる。
```C#
Discreate = new int[2];
Discreate[0] = 0 //上移動
Discreate[1] = 1 //下移動 などなど


//例：OnActionReceived()内では、引数のactionBuffers.DiscreteActions[0]に、0か1が入っている。
//0なら上移動、1なら下移動をする。
pubic override void OnActionReceived(ActionBuffers actionBuffers)
{
    int action = actionBuffers.DiscreteActions[0]; 
    //DiscreteActionsの配列サイズは、BehaviorParameters > Discrete Branchesの値で決まる。
}

```
* Unity側のパラメータ設定を、行動サイズに合わせる。
    * 

* 連続値の行動と離散値の行動において、アルゴリズムごとにbatch_sizeなどの典型的な値の範囲が変わるので注意
    * 

### 4-3 Vector Observation

カメラやレンダリングされたテクスチャを観察として利用する機能。
* グレースケールで観測する場合は、2次元配列に
* カラーで観測する場合は、3次元配列になる。





### 4-4 RayCast Observation

カメラ型の観察とは異なり、エージェントからレーザー線を飛ばし、そのレーザー線が衝突したオブジェクトの距離を観察することで、観察を行う方法.
* Unity側のパラメータ設定で、検出対象のオブジェクトを複数個設定することができる。
* レイとオブジェクトが衝突した場合は、ヒット：１, ミス：０, 距離：`float`
* 衝突しない場合は、ヒット：０, ミス：１, 距離：0





### 4-5 セルフプレイ

対戦ゲームのこと。一方の勝利が他方の損失になるようなゲームのこと。
* より具体的には、現在のポリシー VS  過去に学習したポリシー
これを行うことで、他のポリシーに対して行動意思決定を強化することができ、様々なポリシーに対応できるようになる。
